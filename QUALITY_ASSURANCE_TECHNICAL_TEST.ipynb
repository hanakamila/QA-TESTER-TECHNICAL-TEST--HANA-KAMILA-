```python
# TEST CASE SCN_01_TC_01

# read the table
import pandas as pd
booking = pd.read_excel('/content/booking.xlsx')

# find the duplicates booking based on venue, date, start time, and end time
duplicates = booking[
    booking.duplicated(
        subset=["venue_id", "date", "start_time", "end_time"],
        keep=False
    )
]

# print reports on duplicate booking found
print("===== DUPLICATE BOOKING REPORT =====")

if duplicates.empty:
    print("no duplicate bookings found.")
else:
    print(f"found {len(duplicates)} duplicate booking(s):\n")
    print(
        duplicates[
            [
                "Booking_id",
                "venue_id",
                "date",
                "start_time",
                "end_time",
                "price",
            ]
        ]
    )
```

# TEST CASE SCN_01_TC_02

```python
# TEST CASE SCN_01_TC_02

# read the table
import pandas as pd
booking = pd.read_excel('/content/booking.xlsx')
schedule = pd.read_excel('/content/schedule.xlsx')

# find incorrect price on booking table based on schedule table
comparison = booking.merge(
    schedule,
    on=["venue_id", "date", "start_time", "end_time"],
    how="left",
    suffixes=("_booking", "_schedule")
)

wrong_price = comparison[
    comparison["price_booking"] != comparison["price_schedule"]
]

# print rows that has incorrect prices
print("===== INCORRECT PRICE REPORT =====")

if wrong_price.empty:
    print("no incorrect price bookings found.")
else:
    print(f"found {len(wrong_price)} incorrect price booking(s):\n")
    for index, row in wrong_price.iterrows():
        print(f"Booking ID     : {row['Booking_id']}")
        print(f"Venue ID       : {row['venue_id']}")
        print(f"Date           : {row['date']}")
        print(f"Start Time     : {row['start_time']}")
        print(f"End Time       : {row['end_time']}")
        print(f"Booking Price  : Rp {row['price_booking']:,}")
        print(f"Correct Price  : Rp {row['price_schedule']:,}")
        print(f"Difference     : Rp {row['price_booking'] - row['price_schedule']:,}")
```
